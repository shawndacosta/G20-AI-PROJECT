```python
import os
import pickle
import pandas as pd
import random

base = r"C:\Users\camil\OneDrive\Bureau\house_prediction"

DATA_PATH = os.path.join(base, "data", "cleaned", "cleaned_house.csv")
MODEL_PATH = os.path.join(base, "models", "house_best_model.pkl")

df = pd.read_csv(DATA_PATH)
print(f"Dataset loaded: {len(df)} rows")

with open(MODEL_PATH, "rb") as f:
    model_data = pickle.load(f)

model = model_data["model"]
features = model_data["features"]

print("Model loaded. Features:", features, "\n")

samples = df.sample(10, random_state=random.randint(0, 999999))

for idx, row in samples.iterrows():
    X = pd.DataFrame([row[features]])
    pred = model.predict(X)[0]

    print("="*80)
    print("HOUSE FEATURES")
    print(row[features])
    print("\nREAL PRICE :", row["price"])
    print("PREDICTED  :", int(pred))
```
